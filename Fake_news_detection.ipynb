```python
data_merge=pd.concat([data_fake,data_true],axis=0)
data_merge.head(10)
```

```python
data_merge.columns
```

```python
data = data_merge.drop(['title','subject', 'date'], axis=1)
```

```python
data.isnull().sum()
```

```python
data=data.sample(frac=1)
```

```python
data.head()
```

```python
data.reset_index(inplace=True)
data.drop(['index'],axis=1,inplace=True)
```

```python
data.columns
```

```python
data.head()
```

```python
def wordopt(text):
    text=text.lower()
    text=re.sub('\[.*?\]','',text)
    text=re.sub("\\W"," ",text)
    text=re.sub('https?://\S+|www\.\S+','',text)
    text=re.sub('<.*?>+','',text)
    text=re.sub('[%s]'%re.escape(string.punctuation),'',text)
    text=re.sub('\n','',text)
    return text
```

```python
data['text']=data['text'].apply(wordopt)
```

```python
x=data['text']
y=data['class']
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
vectorization=TfidfVectorizer()
xv_train=vectorization.fit_transform(x_train)
xv_test=vectorization.transform(x_test)
```

```python
from sklearn.linear_model import LogisticRegression
LR=LogisticRegression()
LR.fit(xv_train,y_train)
```

```python
pred_lr=LR.predict(xv_test)
```

```python
LR.score(xv_test,y_test)
```

```python
print(classification_report(y_test,pred_lr))
```

```python
from sklearn.tree import DecisionTreeClassifier
DT=DecisionTreeClassifier()
DT.fit(xv_train,y_train)
```

```python
pred_dt=DT.predict(xv_test)
```

```python
DT.score(xv_test,y_test)
```

```python
print(classification_report(y_test,pred_lr))
```

```python
from sklearn.ensemble import GradientBoostingClassifier
GB=GradientBoostingClassifier(random_state=0)
GB.fit(xv_train,y_train)
```

```python
predit_gb=GB.predict(xv_test)
```

```python
GB.score(xv_test,y_test)
```

```python
print(classification_report(y_test,predit_gb))
```

```python
from sklearn.ensemble import RandomForestClassifier
RF=RandomForestClassifier(random_state=0)
RF.fit(xv_train,y_train)
```

```python
pred_rf=RF.predict(xv_test)
```

```python
RF.score(xv_test,y_test)
```

```python
print(classification_report(y_test,pred_rf))
```

```python
def output_label(n):
    if n == 0:
        return "Fake News"
    elif n == 1:
        return "True News"

def data_fake_manual_testing(news):
    testing_news = {"text": [news]}
    new_def_test = pd.DataFrame(testing_news)
    new_def_test["text"] = new_def_test["text"].apply(wordopt)   # wordopt ফাংশন তোমার preprocessing
    new_def_test = vectorization.transform(new_def_test["text"])

    pred_LR = LR.predict(new_def_test)
    pred_DT = DT.predict(new_def_test)
    pred_GB = GB.predict(new_def_test)
    pred_RF = RF.predict(new_def_test)

    print("\n\nLR Prediction: {}"
          "\nDT Prediction: {}"
          "\nGB Prediction: {}"
          "\nRF Prediction: {}".format(
              output_label(pred_LR[0]),
              output_label(pred_DT[0]),
              output_label(pred_GB[0]),
              output_label(pred_RF[0])
          ))
```
